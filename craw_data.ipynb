```python
def sum_arr ( nums):
    if len(nums) ==0:
        return []
    sum_nums = [nums[0]]
    for i in range(1,len(nums)):
        temp_sum = nums[i]+sum_nums[i-1]
        sum_nums.append(temp_sum)
    return sum_nums
```

```python
a = [3,1,2,10,1]
b = sum_arr(a)
print(b)
```

```python
a = [0]
b = sum_arr(a)
print(b)
```

```python
def minimum_operator(n):
    result = 0
    if n%2 ==1:
        l = n- n//2*2 - 1
        r = n- n//2*2 + 1
    else:
        l = n/2 -1
        r = n/2 
    for i in range(n//2):
        temp = (2*l+1 )/2 -(2*r+1)/2
        if temp <0:
            temp = -temp
        result += temp
        l-=1
        r+=1
    return result
```

```python
a = minimum_operator(10)
print(a)
```
